```python
from commonfunctions import *
from skimage.filters import gaussian
from skimage.filters import laplace

import math

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
'''

(1) Averaging Filter
'''
img=rgb2gray(io.imread('pout.tif'))
gausnoise=random_noise(img,mode='gaussian',var=0.01)
afterfilter=gaussian(gausnoise,sigma=1)
show_images([img,gausnoise,afterfilter],['original image','gaussian noise','after averaging filter'])
```

```python
'''
(2) Median Filter
'''
img=rgb2gray(io.imread('tire.tif'))
spnoise=random_noise(img,mode='s&p',amount=0.05)
afterfilter=median(spnoise)
show_images([img,spnoise,afterfilter],['original image','sp noise','after averaging filter'])

```

```python
'''
(BONUS)
1-Make edge detection
2-Multiply by factor < 1
3-Subtract (Orginal Image) - (Edge Detection Image)
'''

img=rgb2gray(io.imread('pout.tif'))
edges=sobel(img)
edges=edges*0.9
aftersub=(img/256)-edges
show_images([img,edges,aftersub],['original image','edges','after subtraction'])


```
