```python
# 1. Built-in Sobel, Prewill, Roberts

img = rgb2gray(io.imread('circuit.tif'))
mask=np.ones(img.shape)
imprewit=prewitt(img,mask) #mask can be provided 
imroberts=roberts(img)
imsobel=sobel(img)
imsobel_h=sobel_h(img)
imsobel_v=sobel_v(img)
imcanny=canny(img,sigma=1,low_threshold=20,high_threshold=60)
imcanny1=canny(img,sigma=2.5,low_threshold=20,high_threshold=60)
imcanny2=canny(img,sigma=1,low_threshold=20,high_threshold=150)
#Sobel operator is simple, but its accuracy suffers in noisy conditions. 
#On the contrary, Canny edge detector has many favorable features such as smoothing effect to remove noise
#, and improving signal to noise ratio through a process known as non-maximal suppression.
show_images([img,imprewit,imroberts,imsobel,imsobel_h,imsobel_v,imcanny],['original image','prewit','roberts','sobel','sobel_h','sobel_v','canny'])
show_images([imcanny,imcanny1,imcanny2],['normal canny','canny with higher sigma','canny with high high threshold '])
```

```python
# 2. Custom Sobel (V-H-Both)

def sobelcustom(img,threshold):
    hy=np.array([[-1,0,1], # error in document here 
   [-2,0,2],
   [-1,0,1]])

    hx =np.array([[-1,-2,-1],
    [0 ,0 , 0]
    ,[1 ,2 , 1]])
    imgx=convolve2d(img,hx)
    imgy=convolve2d(img,hy)
    newimg=np.sqrt(imgx**2 + imgy**2)
    newimg[newimg<threshold]=0
    show_images([img,imgx,imgy,newimg],['img','x filter','y filter','x and y together'])
    return newimg


img = rgb2gray(io.imread('circuit.tif'))
sobelimg=sobelcustom(img,20)
```

```python
# 3. Custom LoG
thres1= 20
thre2=30
def logcustom(img,threshold,threshold2):
    f1=np.array([[-1,-1,-1],
   [-1,8,-1],
   [-1,-1,-1]])

    f2 =np.array([[0,1,0],
    [1 ,-4 , 1]
    ,[0 ,1 , 0]])
    afterfilter=gaussian(img,sigma=2.5)*255  #ask how to apply gaussian threshold
    imgf1=convolve2d(afterfilter,f1)
    imgf2=convolve2d(afterfilter,f2)
    imgf1[imgf1<threshold]=1
    imgf2[imgf2<threshold2]=1
    imgf1[imgf1>threshold]=0
    imgf2[imgf2>threshold2]=0
    show_images([afterfilter,imgf1,imgf2],['img after filter','f1 filter','f2 filter'])
    return 0

img = rgb2gray(io.imread('circuit.tif'))
logimg=logcustom(img,3,1)
```
